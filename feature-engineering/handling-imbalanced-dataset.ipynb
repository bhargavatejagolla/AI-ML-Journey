exmaple of imablanced data set is : 9:1 ratio of classes, where one class has 90% of the samples and the other class has only 10%. This can lead to biased models that perform well on the majority class but poorly on the minority class. To address this issue, techniques such as oversampling the minority class, undersampling the majority class, or using algorithms that are designed to handle imbalanced data can be employed.

#### we have two types of sampling that is
1.Up sampling<br>
2.Down sampling<br>
 

in upsampling we try to increase the data points from the lower one .In our exampel from the 10% .<br>
Whereas in downsampling we try to decrease the data points from the higher one .In our example from the 90% .<br>

```python
import numpy as np 
import pandas as pd
np.random.seed(123)

n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples -n_class_0
```

```python
n_class_0, n_class_1    
```

```python
## create my dataframe with imbalanced target variable
class_0 = pd.DataFrame({
    'feature_1':np.random.normal(loc =0,scale =1,size = n_class_0),
    'feature_2':np.random.normal(loc = 0,scale = 1,size = n_class_0),
    'target':[0]*n_class_0
})
```

```python

class_1 = pd.DataFrame({
    'feature_1':np.random.normal(loc = 2,scale = 1,size = n_class_1),
    'feature_2':np.random.normal(loc = 2,scale = 1,size = n_class_1),
    'target':[1]*n_class_1
})
```

```python
df =pd.concat([class_0,class_1]).reset_index(drop = True)
```

```python
df['target'].value_counts()
```

```python
## upsampling 
df_minority =df[df['target']==1]
df_majority = df[df['target']==0]
```

from sklearn.utils import resample
# resampel means it tires to make the minority class same as majority class by randomly sampling with replacement from the minority class

```python
from sklearn.utils import resample
df_minority_upsampled=resample (df_minority ,replace = True,
          n_samples = len(df_majority),
          random_state = 123)
```

```python
df_minority_upsampled.shape
```

```python
df_upsampled = pd.concat([df_majority,df_minority_upsampled]).reset_index(drop = True)
```

```python
df_upsampled['target'].value_counts()
```

why replace = False for down sampling and replace = True for up sampling?
because in down sampling we are reducing the number of samples in the majority class, so we do not want to replace any samples. In up sampling, we are increasing the number of samples in the minority class, so we need to replace some samples to create new ones.
