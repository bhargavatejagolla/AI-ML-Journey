## SMOTE (Synthetic Minority Over Sampling Technique) for handling imbalanced datasets)


SMOTE is technique used in machine learning to address imbalanced datasets.where the minority class has significntly fewer instances than the majority class.SMOTE works by creating synthetic examples of the minority class by interpolating between existing minority class instances and their nearest neighbors in the feature space.This helps to balance the dataset and improve the performance of machine learning models trained on it.

interpolating means creating new data points by taking the average of existing data points. In the context of SMOTE, it involves generating synthetic examples of the minority class by averaging the feature values of existing minority class instances and their nearest neighbors. This helps to create a more balanced dataset for training machine learning models.

`make_classification` is a function in the scikit-learn library that generates a synthetic dataset for classification tasks. It allows you to create a dataset with specified characteristics, such as the number of samples, features, classes, and class separation. This function is often used for testing and evaluating machine learning algorithms on controlled datasets.

```python
from sklearn.datasets import make_classification
x,y =make_classification(n_samples = 1000,n_redundant=0,n_features = 2,n_clusters_per_class = 1,weights=[0.90],random_state = 123)
```

the parameters used in the make_classification function are as follows jsut definitions:
- n_samples: This parameter specifies the total number  of samples to generate in the dataset. In this case, it is set to 1000, meaning that the generated dataset will contain 1000 samples.
- n_redundant: This parameter specifies the number of redundant features to generate. Redundant features are linear combinations of the informative features. In this case, it is set to 0, meaning that no redundant features will be generated.
- n_features: This parameter specifies the total number of features to generate in the dataset. In this case, it is set to 2, meaning that the generated dataset will contain 2 features.
- n_clusters_per_class: This parameter specifies the number of clusters to generate for each class.
In this case, it is set to 1, meaning that each class will be represented by a single cluster of samples.
- weights: This parameter specifies the weights for each class in the dataset. It is a list
that indicates the proportion of samples for each class. In this case, it is set to [0.90], meaning that 90% of the samples will belong to one class (the majority class) and the remaining 10% will belong to the other class (the minority class).
- random_state: This parameter is used to control the randomness of the generated dataset. It is an integer value that serves as a seed for the random number generator. Setting a specific random_state ensures that the same dataset is generated each time the code is run, which can be useful for reproducibility. In this case, it is set to 123.


```python
import pandas as pd
df1 = pd.DataFrame(x,columns =['f1','f2'])
df2 = pd.DataFrame(y,columns = ['target'])
df = pd.concat([df1,df2],axis = 1)
df.head()
```

```python
df["target"].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["f1"],df["f2"],c = df["target"])
```
