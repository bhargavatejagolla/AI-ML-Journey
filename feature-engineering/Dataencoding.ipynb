### Data Encoding
Data Encoding is the process of converting categorical data into numerical data. This is necessary because machine learning algorithms typically require numerical input. There are several techniques for encoding categorical data, including:

- Nominal/OHE Encoding
- Label and Ordinal Encoding
- Target Guided Ordinal Encoding
- Binary Encoding

#### Nominal/OHE Encoding
One hot encoding , also known as nominal encoding is technique to represent categorical data as numerical data,whcih is more suitable for machine learning algorithms.In this Technique,each category is represented as binary vector,where each bit corresponds to a unique category.For example,if we have a categorical feature "Color" with three categories:Red,Green and Blue,one hot encoding would represent these categories as follows:


Red:[1,0,0] Green:[0,1,0] Blue:[0,0,1]

#### disadvantages
-1. It is not a good method to handle outliers
-2. It is not a good method to handle skewed data
-3. It is not a good method to handle missing values
-4. It is not a good method to handle categorical data
-5. It is not a good method to handle high cardinality data

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
##create a simpel data frame
df = pd.DataFrame({"color": ["red", "blue", "green","green", "red", "blue"]})
print(df)
```

```python
### create an instance of one hot encoder
encoder = OneHotEncoder()
```

```python
## perform fit and transform
# fit_tansform returns a sparse matrix. To convert it to a dense array, we can use the toarray() method.
# fit_transform means that we are fitting the encoder to the data and then transforming it in one step. This is a common practice when we want to encode the data and use it for machine learning models.

encoded =encoder.fit_transform(df[["color"]]).toarray()
```

```python
import pandas as pd
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(["color"]))
```

## Label Encoding
Label encoding  and ordinal encoding are two techniques used to encode categorical data as numerical data.

Label encoding involves assigning a unique numerical label to each category in the variable.The labels are usually assigned in alphabetical order or based on the frequency of the categories.For example, if we have a variable "color" with categories "red", "blue", and "green", we can assign the labels as follows:

1. Red :1
2. Green : 2
3. Blue : 3

```python
df = pd.DataFrame({"color": ["red", "blue", "green", "green", "red", "blue"]})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbl_encoder = LabelEncoder()
encoded_lbl= lbl_encoder.fit_transform(df["color"])
encoded_lbl
```

```python
lbl_encoder.transform(["red"])
```

## the problem with label encoding is that it assigns an arbitrary numerical value to each category, which can lead to unintended consequences when used in machine learning models. For example, if we have three categories "red", "blue", and "green", label encoding might assign 0 to "red", 1 to "blue", and 2 to "green". This implies an ordinal relationship between the categories, which may not be the case. In contrast, one-hot encoding creates binary columns for each category, which avoids this issue and allows the model to treat each category independently.

### Ordinal Encoding

It is used to encode categorical data that have an intrinsic order or ranking.That means that the categories have a meaningful order or hierarchy. For example, if we have a feature called "education level" with categories like "high school", "bachelor's degree", "master's degree", and "PhD", we can use label encoding to assign numerical values to these categories based on their order. In this case, we might assign 0 to "high school", 1 to "bachelor's degree", 2 to "master's degree", and 3 to "PhD". This way, the encoded values reflect the ordinal relationship between the categories.
based on its position in the order.For example,if we haev categorical variables "education level " with four possible values("high school", "bachelor's degree", "master's degree", and "PhD"), label encoding would assign the following numerical values:

1.High School
2.College
3.Graduate
4.PhD

Here the ranks are important.

```python
from sklearn.preprocessing import OrdinalEncoder

```

```python
df = pd.DataFrame({
    "size":["small","medium","large","medium","small","large"]
})
```

```python
df
```

```python
##create an instance of ordinal encoder
ordinal_encoder = OrdinalEncoder(categories=[["small","medium","large"]])
encoded_ordinal = ordinal_encoder.fit_transform(df[["size"]])
```

## Target Guided Ordinal Encoding

It is a technique used to encode categorical variables into numerical values. This is often necessary because many machine learning algorithms require numerical input.

This encoding technique is useful when we have a categorical variabel with a large number of unique categories and we want to use this variable as a feature in our machine learning model

In Target Guided Ordinal Encoding,we replace each category with a numerical value based on the mean or median of the taregt variable for that category.This creates a monotonic relationship between the categorical variable and the target variable, which can be beneficial for certain machine learning algorithms. This encoding technique is particularly useful when there is a natural order or hierarchy among the categories, and it can help capture the relationship between the categorical variable and the target variable more effectively than other encoding methods.

```python
import pandas as pd
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix',"New York"],
    'temperature': [30, 25, 20, 35, 40,60]
})
```

```python
df
```

```python
mean_temp = df.groupby("city")["temperature"].mean().to_dict()
```

```python
df['City_encoded'] = df["city"].map(mean_temp)
```
