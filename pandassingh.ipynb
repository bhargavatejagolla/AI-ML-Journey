## Pandas-DataFrame And Series

Pandas is a powerful data manipulation library in Python ,widely used for data analysis and data cleaning .It provides two primary data structures :Series and DataFrame.A Series is a one-dimensional array -like object,while a Data Frame isa two-dimensional ,size -mutable ,and potentially heterogeneous tabular data structure with labeled axes (rows and columns)

```python
import pandas as pd
```

```python
## Series 
###  A pandas series is a one dimensional array -like object that can hold any data type : It is similar to 
import pandas as pd 
data = [1,2,3,4,5]
series = pd.Series(data)
print("series \n",series)
```

```python
## create a series from dictionary 
data = {'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(series_dict)
```
