```python
import sqlite3
```

```python
connection = sqlite3.connect('example.db')
```

```python
connection
```

```python
#it is used to iterate over the tables and to retrieve info
cursor = connection.cursor()
```

```python
#create a table
cursor.execute('''
Create  Table If Not Exists  employees (
    id integer primary key,
    name Text Not NUll,
    age Integer,
    department text)
    ''')
##commit the changes
connection.commit() 
```

```python
cursor.execute(
    '''
    select * from employees
    '''
)
```

```python
## Insert the data in sqlite table
cursor.execute('''
Insert into employees(name ,age ,department)
                values('bt',19,'Data Scientist')''')
cursor.execute('''
Insert into employees(name ,age ,department)
                values('champak',32,'Data engineer')''')

cursor.execute('''
Insert into employees(name ,age ,department)
                values('kamash',19,'champar')''')
connection.commit()
```

```python
#Quering the data from the table
cursor.execute(
    '''
    select * from employees
    '''
)
rows = cursor.fetchall()
##printt the queried data
for row in rows:
    print(row)
```

```python
#update the data in the table
cursor.execute('''
UPDATE employees
set age = 34 
where name = 'bt' 
''')
connection.commit()
```

```python
# Quering the data from the table
cursor.execute(
    '''
    select * from employees
    '''
)
rows = cursor.fetchall()
# printt the queried data
for row in rows:
    print(row)
```
