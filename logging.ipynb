# Python Logging

Logging is a crucial aspect of any application ,providing a way to track events ,errors,and operational information.Python's built-in logging module offers a flexible framework for emitting log messages from Python programs.In this lesson,we will cover the basics of logging ,including how to configure logging,log levels and best practices for using logging in Python application

```python
import logging
```

```python
## configure the basic logging settings
logging.basicConfig(level=logging.DEBUG)

## log messages 
logging.debug("this is a debug message")
logging.info("this is an info message")
logging.warning("this is a warning message")
logging.error("this is an error message")
logging.critical("this is a critical message")
```

## Log Levels

Python's logging module has several log levels indicating the severity of events.The default levels are:

<li>DEBUG:Detailed information ,typically of interest only when diagnosing problems.</li><br>
<li>INFO:Confirmation that things are working as expected</li><br>
<li>WARNING:An Indication that something unexpected happened or indicative of some problem in the near future (eg.. 'disk space low') .The software is till working as expected</li><br>
<li>ERROR:Due to a more serious problem,the software has not been able to perform some function.</li><br>
<li>CRITICAL:A very serious error,indicating that the program itself may be unable to continue runnning</li>

```python
## Configuring the logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level = logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
    
    
)

# configure the basic logging settings


# log messages
logging.debug("this is a debug message")
logging.info("this is an info message")
logging.warning("this is a warning message")
logging.error("this is an error message")
logging.critical("this is a critical message")
```

# Logging with real world Example

```python
import logging 
# create a logger for module1 
logger1 = logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

#create a loggger for module2 
logger2  =logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

logging.basicConfig(
    level = logging.DEBUG,
    format = '%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt = '%Y-%m-%d %H:%M:%S'
)
```

```python
#log message with different logger
logger1.debug("This is a debug message for module1")
logger2.warning("This is a warning message for module2")
logger2.error("This is an error message")
```
