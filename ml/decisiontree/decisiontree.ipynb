```python
import pandas as  pd
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
X= pd.DataFrame(iris['data'], columns = ['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm'])
```

```python
Y = iris['target']
```

```python
### tain test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state= 10)
```

```python
## apply decision tree classifier 
from sklearn.tree import DecisionTreeClassifier
tree_classifier = DecisionTreeClassifier()
```

```python
tree_classifier.fit(X_train,Y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier,filled = True)
```

### post pruning is set or sued with the help of max_depth 

```python
# apply decision tree classifier
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
tree_classifier2 = DecisionTreeClassifier(max_depth=2)
tree_classifier2.fit(X_train, Y_train)
plt.figure(figsize=(15, 10))
tree.plot_tree(tree_classifier2, filled = True)
```

```python
y_pred1 = tree_classifier.predict(X_test)
y_pred2 = tree_classifier2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm1 = confusion_matrix(Y_test,y_pred1)
clf1 = classification_report(Y_test,y_pred1)
print(cm1)
print(clf1)
```

```python
cm2 = confusion_matrix(Y_test, y_pred2)
clf2 = classification_report(Y_test, y_pred2)
print(cm2)
print(clf2)
```

### Decision Tree Classifier and Hyper parameter tuning

```python
params = {
    'criterion' :["gini","entropy","log_loss"],
    'splitter' :['best','random'],
    'max_depth':[1,2,3,4],
    'max_features':['auto','sqrt','log2']
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
tree_model = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(tree_model,param_grid=params,cv = 5,scoring  = 'accuracy')
```

```python
grid.fit(X_train,Y_train)
```

```python
pred3 = grid.predict(X_test)
```

```python
cm1 = confusion_matrix(Y_test,  pred3)
clf1 = classification_report(Y_test, pred3)
print(cm1)
print(clf1)
```
