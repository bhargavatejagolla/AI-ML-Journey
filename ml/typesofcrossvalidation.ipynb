# Cross Validation
- ### Types of Cross Validation
- #### Leave One out Cross Validation
- #### Leave P out Cross Validation
- #### Stratified K-Fold Cross Validation
- #### Time Series Cross Validation



 Leave  one out Cross Validation can be defined as in deatil that  is used to evaluate the performance of a machine learning model. In this technique, we use one data point as the test set and the remaining data points as the training set. This process is repeated for each data point in the dataset, resulting in multiple iterations of training and testing. 

### disadvantages of Leave one out cross validation
1. `Computationally expensive`: It requires training the model n times, where n is the number of data points, which can be time-consuming for large datasets.
2. `High variance`: Since each training set is almost identical to the original dataset, the model's performance can vary significantly across different iterations, leading to high variance in the evaluation results.
3. `Not suitable for small datasets`: If the dataset is small, leaving out one data point can significantly affect the training process, leading to unreliable estimates of model performance.
4. `may be overfitting`


Leave P out Cross validation can be defined as the process of dividing the dataset into P subsets, where P is a positive integer less than the total number of samples in the dataset. The model is trained on P-1 subsets and tested on the remaining subset. This process is repeated P times, with each subset serving as the test set once. The performance of the model is then averaged over all P iterations to obtain an overall estimate of its performance.

K fold Cross validation can be defined as a resampling procedure used to evaluate machine learning models on a limited data sample. The procedure has a single parameter called k that refers to the number of groups that a given data sample is to be split into. As such, the procedure is often called k-fold cross-validation. Stratified K Fold Cross Validation can be defined as a variation of K Fold Cross Validation. In Stratified K Fold Cross Validation, the data is split in such a way that each fold has the same proportion of classes as the original dataset. This is particularly useful when dealing with imbalanced datasets, where some classes may be underrepresented. By ensuring that each fold has a representative distribution of classes, Stratified K Fold Cross Validation provides a more accurate evaluation of the model's performance across all classes.

Times Series Cross Validation can be defined as a method of evaluating the performance of a machine learning model on time series data. It involves splitting the data into multiple training and testing sets, where the training set is used to train the model and the testing set is used to evaluate its performance. The key difference between time series cross validation and traditional cross validation is that in time series cross validation, the training and testing sets are created in a way that respects the temporal order of the data. This means that the training set will always contain data that comes before the testing set, which helps to prevent data leakage and ensures that the model is evaluated on unseen data.
